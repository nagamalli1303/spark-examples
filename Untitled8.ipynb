```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import Dropout
```

```python
import pandas as pd
healthCareDf = pd.read_csv('/content/HealthCare_Customer_Reviews.csv').sample(n=20000)

print(healthCareDf.shape)
```

```python
healthCareDf.head(20)
```

```python
Score = []
for row in healthCareDf['Rating']:
  if row < 3 :
    Score.append(0)
  else:
    Score.append(1)
```

```python
healthCareDf["Score"]=Score
healthCareDf.head(5)
```

```python
healthCareDf=healthCareDf.dropna()
X=healthCareDf.drop('Score',axis=1)
```

```python
y=healthCareDf['Score']
```

```python

messages=X.copy()
```

```python
messages.reset_index(inplace=True)
messages["Content"][0]
```

```python
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    #print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['Content'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus[0]
```

```python
voc_size=5000
onehot_repr=[one_hot(words,voc_size)for words in corpus]
onehot_repr[0]
```

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
## Creating model
embedding_vector_features=40
model1=Sequential()
model1.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model1.add(Bidirectional(LSTM(100)))
model1.add(Dropout(0.3))
model1.add(Dense(1,activation='sigmoid'))
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model1.summary())
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train1, X_test, y_train1, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train1, y_train1, test_size=0.33, random_state=42)
```

```python
model1.fit(X_train,y_train,validation_data=(X_val,y_val),epochs=10,batch_size=64)
```

```python
X_test[0]
```

```python
from itertools import chain
y_pred=(model1.predict(X_test) > 0.5).astype("int32")

y_predict = [ele for sublist in y_pred for ele in sublist]
```

```python
print(y_predict)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)
```
